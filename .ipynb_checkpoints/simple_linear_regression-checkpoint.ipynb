```python
# download dataset
import wget
datasetName = wget.download("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/FuelConsumptionCo2.csv")
```

```python
print(datasetName)
```

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

# read dataset into df
df = pd.read_csv(datasetName)

# peek at the dataset
# df.head()

# select some features for regression
cdf = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']]

# peek at the selection
# cdf.head()

# plot emission values with respect to the engine size
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color = 'blue')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Emission Values VS Engine Size')
plt.show()
```

```python
# perform train/test split
# randomly select 80% of the observations as the train dataset
# ... and the remaining 20% as the test dataset using a mask
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]

# check the training data distribution
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color = 'green')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Training Data Distribution')
plt.show()
```

```python
# create a linear regression model
from sklearn import linear_model
regr = linear_model.LinearRegression()

# independent variable: Engine Size
train_x = np.asanyarray(train[['ENGINESIZE']])

# dependent variable: CO2 Emissions
train_y = np.asanyarray(train[['CO2EMISSIONS']])

# fit x and y into a fitting line
regr.fit(train_x, train_y)

# Theta 0
intercept = regr.intercept_[0]
print('The Intercept: ', intercept)
# Theta 1
coef = regr.coef_[0][0]
print('The Coefficient: ', coef)
```

```python
# plot the fitting line using the following call signature
# plt.plot(x, y, 'r-')
# where y is the dependent variable
# y = theta0 + theta1 * x
plt.plot(train_x, intercept + coef * train_x, 'r-')
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color = 'green')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Training Data Distribution with the Fit line')
plt.show()
```

```python
# predict CO2 Emissions for the test dataset
test_x = np.asanyarray(test[['ENGINESIZE']])
# predict y (i.e. the dependent variable) ...
# given x (i.e. the independent variable) ...
# using regr (i.e. the regression model built above)
test_y = regr.predict(test_x)
i = 0
for xValue in test_x:
    print('For Engine Size ', xValue, ', Predicted CO2 Emission: ', test_y[i])
    i = i + 1
#   print the first 10 predictions
    if i >= 10:
        break
```

```python
# evaluation of prediction
# best possible r2 score is 1
from sklearn.metrics import r2_score
# absolute CO2 Emission values from the test dataset, not predictions
test_y_abs = np.asanyarray(test[['CO2EMISSIONS']])

print('Absolute Mean Error: %.2f' %np.mean(np.abs(test_y - test_y_abs)))
print('Residual Sum of Squares: %.2f' %np.mean((test_y - test_y_abs) ** 2))
print('R2-Score: %.2f' %r2_score(test_y, test_y_abs))
```
